```python
import pandas as pd
import numpy as np
```

```python
housing=pd.read_csv('train.csv')
housing.head()
```

```python
lables = housing[['SalePrice']]
features =housing.drop(['SalePrice'], axis=1)
```

```python
features.head()
```

```python
housing.info()
```

```python
lables.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(features, lables, test_size=0.2,random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder

# Pipeline for numeric attributes
numeric_pipeline = Pipeline([
    ('median_imputer', SimpleImputer(strategy='median')),
    ('std_scaler', StandardScaler())
])

# Pipline for categorical attributes
categorical_pipeline = Pipeline([
    ('one_hot_encoder',OneHotEncoder())
])
```

```python
from sklearn import set_config

set_config(display='diagram')

numeric_pipeline
```

```python
from sklearn.compose import ColumnTransformer

numeric_attributes = [
    'Id','MSSubClass','OverallQual','OverallCond','YearBuilt','TotalBsmtSF','LowQualFinSF','GrLivArea','Fireplaces','GarageCars','GarageArea','WoodDeckSF','OpenPorchSF','ScreenPorch','PoolArea','LotArea','MoSold','YrSold',
    
]

pipeline = ColumnTransformer([
    ('numerical_transformers', numeric_pipeline, numeric_attributes),
    ('categorical_transformers', categorical_pipeline, ['Neighborhood']),
])
```

```python
pipeline
```

```python
pipeline.fit_transform(x_train)[0]
```

```python
from sklearn.linear_model import LinearRegression

model_pipline = Pipeline([
    ('preprocessing', pipeline),
    ('model', LinearRegression())
    
])

model_pipline
```

```python
model_pipline.fit(x_train, y_train)
```

```python
model_pipline.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

predictions = model_pipline.predict(x_test)
rmse = mean_squared_error(y_test, predictions, squared=False)
r2 = r2_score(y_test, predictions)
print(f'RMSE: {rmse}')
print(f'R-squared: {r2}')
```
